$$C_t= C_1n_t^{\alpha{}}e^{u_t}$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
C = [100]
production = 300
alpha = -0.05
for n in range(1, production):
    ut = np.random.normal(0, 0.05)
    costs = C[0] * n**alpha * np.e**ut
    C.append(costs)
plt.plot(range(1, production+1), C)
```

```python
for sigma in range(10,5,-1):
    C = [100]
    production = 100
    alpha = -0.1
    for n in range(1, production):
        ut = np.random.normal(0, sigma/100)
        costs = C[0] * n**alpha * np.e**ut
        C.append(costs)
    plt.plot(range(1, production+1), C)
```
