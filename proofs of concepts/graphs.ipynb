## Learning curve

$$C_t= C_1n_t^{\alpha{}}e^{u_t}$$

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
C = [1000]
production = 300
alpha = -0.05
for n in range(1, production):
    ut = np.random.normal(0, 0.025)
    costs = C[0] * n**alpha * np.e**ut
    C.append(costs)
plt.plot(range(1, production+1), C)
```

```python
for sigma in range(10,5,-1):
    C = [100]
    production = 100
    alpha = -0.1
    for n in range(1, production):
        ut = np.random.normal(0, sigma/100)
        costs = C[0] * n**alpha * np.e**ut
        C.append(costs)
    plt.plot(range(1, production+1), C)
```

```python
def colorFader(c1,c2,mix=0): #fade (linear interpolate) from color c1 (at mix=0) to c2 (mix=1)
    c1=np.array(mpl.colors.to_rgb(c1))
    c2=np.array(mpl.colors.to_rgb(c2))
    return mpl.colors.to_hex((1-mix)*c1 + mix*c2)

def closenessToProfit(profit, optimal_profit):
    if profit <= 0:
        return 0
    return (profit / optimal_profit)

profits = []
for q in range(0,101):
    p = 100 - q
    profits.append(p * q - 50*q)

c1='red' #blue
c2='green' #green
n=100

for x in range(0,n+1):
    coef = closenessToProfit(profits[x], max(profits))
    plt.axvline(x, color=colorFader(c1,c2,coef), linewidth=4)
    # print(coef) 

plt.plot(range(0, 101), profits, label='Profits', color='black')

#horizontal line at 0
plt.axhline(0, color='black', lw=0.5, ls='--')
```

```python
# spiral_plot.py

import numpy as np
import matplotlib.pyplot as plt

# Set the range for t (angle, in radians)
t = np.linspace(0, 6 * np.pi, 1000)  # 3 full turns

# Archimedean spiral: r = a + b * t
a = 0
b = 0.2
r = a + b * t

# Convert polar to Cartesian coordinates
x = r * np.cos(t)
y = r * np.sin(t)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.title("Archimedean Spiral")
plt.axis('equal')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
